```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("heart.csv")

# Display the distribution 
plt.figure(figsize=(12, 6))
sns.boxplot(data=dataset.drop(columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']))
plt.title('Distribution of Quantitative Features')
plt.show()

# Detect and remove outliers using the interquartile range (IQR)
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3 - Q1
filtered_dataset = dataset[~((dataset < (Q1 - 1.5 * IQR)) | (dataset > (Q3 + 1.5 * IQR))).any(axis=1)]

# Display the distribution of quantitative features after removing outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_dataset.drop(columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']))
plt.title('Distribution of Quantitative Features after Removing Outliers')
plt.show()
```

```python

# Display a heatmap representing the correlation between features
plt.figure(figsize=(12, 8))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()


```

```python
# Display the density distribution of the dataset for diseased and non-diseased patients
plt.figure(figsize=(12, 6))
sns.kdeplot(data=dataset, x='age', hue='target', common_norm=False, fill=True, palette='husl')
plt.title('Density Distribution of Age for Diseased and Non-Diseased Patients')
plt.xlabel('Age')
plt.show()
```
